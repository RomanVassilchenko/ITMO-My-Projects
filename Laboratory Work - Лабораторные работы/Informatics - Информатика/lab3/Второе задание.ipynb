–ü—Ä–∏–º–µ—Ä:
–í–≤–æ–¥ –í—ã–≤–æ–¥
–í–≤–æ–¥: –ê —Ç—ã –∑–Ω–∞–ª, —á—Ç–æ –í–¢ ‚Äì –ª—É—á—à–∞—è –∫–∞—Ñ–µ–¥—Ä–∞ –≤ –ò–¢–ú–û?
–í—ã–≤–æ–¥: –í–¢ - –ª—É—á—à–∞—è –∫–∞—Ñ–µ–¥—Ä–∞ –≤ –ò–¢–ú–û

```python
import re
```

```python
regexp = r"–í–¢((\s)?(‚Äì)?(\w{0,})?([–∞-—è–ê-–Ø]{0,})?(\s)?){0,4}–ò–¢–ú–û"
def testRegex(value):
    match = re.search(regexp, value)
    if match != None:
        print(match.group())
    else:
        print("None")
```

```python
testRegex("–ê —Ç—ã –∑–Ω–∞–ª, —á—Ç–æ –í–¢ ‚Äì –ª—É—á—à–∞—è –∫–∞—Ñ–µ–¥—Ä–∞ –≤ –ò–¢–ú–û?") #–í–¢ - –ª—É—á—à–∞—è –∫–∞—Ñ–µ–¥—Ä–∞ –≤ –ò–¢–ú–û
testRegex("–í–¢ –ø–æ–ª—É—á–∏—Ç –ø–µ—Ä–≤–æ–µ –º–µ—Å—Ç–æ –≤ —Ç–µ—Å—Ç–µ –ò–¢–ú–û") #None
testRegex("–ò –∫–æ–Ω–µ—á–Ω–æ –∂–µ –≤—Å–µ –≤ –í–¢ –Ω–∞—á–∞–ª–∏ –û–ü–î –∏ –∫–æ–¥–µ–∫—Å –ò–¢–ú–û —É—á–∏—Ç—å") # –í–¢ –Ω–∞—á–∞–ª–∏ –û–ü–î –∏ –∫–æ–¥–µ–∫—Å –ò–¢–ú–û
testRegex("fieiu hfiu3h2ihi2 –í–¢ 3—à–≥23–∏—à - –∫–º–º—É–º—É—É 23 –ò–¢–ú–û –∞—Ü–∞—Ü—É–∞—É") #None
testRegex("fieiu hfiu3h2ihi2 –í–¢ 3—à–≥23–∏—à - –∫–º–º—É–º—É—É 2—É–∞—É–∫ —É–∫–∞—É–∫–ø—É–∫3 –ò–¢–ú–û –∞—Ü–∞—Ü—É–∞—É") #None
```

–î–æ–ø. –∑–∞–¥–∞–Ω–∏–µ No2 (+22 –±–∞–ª–ª–æ–≤)
1) –†–µ–∞–ª–∏–∑—É–π—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º–Ω—ã–π –ø—Ä–æ–¥—É–∫—Ç –Ω–∞ —è–∑—ã–∫–µ Python, –∏—Å–ø–æ–ª—å–∑—É—è —Ä–µ–≥—É–ª—è—Ä–Ω—ã–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è –ø–æ –≤–∞—Ä–∏–∞–Ω—Ç—É, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω–æ–º—É –≤ —Ç–∞–±–ª–∏—Ü–µ.
2) –î–ª—è —Å–≤–æ–µ–π –ø—Ä–æ–≥—Ä–∞–º–º—ã –ø—Ä–∏–¥—É–º–∞–π—Ç–µ –º–∏–Ω–∏–º—É–º 5 —Ç–µ—Å—Ç–æ–≤.
3) –ü—Ä–æ—Ç–µ—Å—Ç–∏—Ä—É–π—Ç–µ —Å–≤–æ—é –ø—Ä–æ–≥—Ä–∞–º–º—É –Ω–∞ —ç—Ç–∏—Ö —Ç–µ—Å—Ç–∞—Ö.

–°—Ç—É–¥–µ–Ω—Ç –í–∞—Å—è –æ—á–µ–Ω—å –ª—é–±–∏—Ç –∫—É—Ä—Å ¬´–ö–æ–º–ø—å—é—Ç–µ—Ä–Ω–∞—è –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç—å¬ª. –û–¥–Ω–∞–∂–¥—ã –í–∞—Å–µ –∑–∞–¥–∞–ª–∏ –¥–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ –∑–∞—à–∏—Ñ—Ä–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ, –ø–µ—Ä–µ–¥–∞–Ω–Ω—ã–µ –≤ —Å–æ–æ–±—â–µ–Ω–∏–µ. –ù–µ–¥–æ–ª–≥–æ –¥—É–º–∞—è, –í–∞—Å—è —Ä–µ—à–∏–ª –∑–∞–º–µ–Ω–∏—Ç—å –≤—Å–µ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é –æ—Ç —ç—Ç–æ–≥–æ —á–∏—Å–ª–∞. –§—É–Ω–∫—Ü–∏—é –æ–Ω –ø—Ä–∏–¥—É–º–∞–ª –Ω–µ —Å–ª–æ–∂–Ω—É—é 3ùë•2 + 5, –≥–¥–µ ùë• ‚àí –∏—Å—Ö–æ–¥–Ω–æ–µ —á–∏—Å–ª–æ. –ü–æ–º–æ–≥–∏—Ç–µ –í–∞—Å–µ —Å –µ–≥–æ –¥–æ–º–∞—à–Ω–∏–º –∑–∞–¥–∞–Ω–∏–µ–º.
–ü—Ä–∏–º–µ—Ä:
–í–≤–æ–¥ –í—ã–≤–æ–¥
20 + 22 = 42 
1205 + 1457 = 5297

```python
def testCrypto(value):
    regexp = r"\d+"
    values = re.findall(regexp, value)
    for i in range(len(values)):
        value = re.sub((values[i]), str(3 * (int(values[i]) ** 2) + 5), value)
    print(value)
```

```python
testCrypto("12 + 34 = 46")  # 437 + 3473 = 6353
testCrypto("18 / 3 = 6")  # 977 / 32 = 113
testCrypto("3 - 2 = 1")  # 32 - 17 = 8
testCrypto("15 * 17 = 255")  # 680 * 872 = 195080
testCrypto("2 ^ 3 = 8")  # 17 ^ 32 = 197
```
